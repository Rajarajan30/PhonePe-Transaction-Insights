# 1. Decoding Transaction Dynamics on PhonePe

```python
#  1. State-wise Transaction Trends
```

```python
query1 = """
SELECT 
    state,
    SUM(transaction_count) AS Total_transactions,
    sum(transaction_amount) as Total_transaction_value
FROM agg_trans
GROUP BY state
ORDER BY Total_transactions desc; 
"""
dft1 = pd.read_sql(query1, mydb)
```

```python
dft1
```

```python
# 2 quarter-wise Transaction Trends

query2 = """
SELECT 
    year,
    quarter,
    SUM(transaction_count) AS Total_transactions,
    SUM(transaction_amount) AS Total_transaction_value,
    ROUND(SUM(transaction_amount) / SUM(transaction_count), 2) AS Avg_transaction_value
FROM agg_trans
GROUP BY year, quarter
ORDER BY year ASC, Avg_transaction_value desc;
"""
dft2 = pd.read_sql(query2, mydb)
```

```python
dft2
```

```python
# 3 Payment Category-wise Trends

query3 = """
SELECT 
    Transaction_Type,
    SUM(transaction_count) AS Total_transactions,
    SUM(transaction_amount) AS Total_transaction_value,
    ROUND(SUM(transaction_amount) / SUM(transaction_count), 2) AS Avg_transaction_value
FROM agg_trans
GROUP BY Transaction_Type
ORDER BY Avg_transaction_value desc;
"""
dft3 = pd.read_sql(query3, mydb)
```

```python
dft3
```

```python
# 4 Payment Category-wise and state wise Trends

query4 = """
SELECT 
    Transaction_Type,
    state,
    SUM(transaction_count) AS Total_transactions,
    SUM(transaction_amount) AS Total_transaction_value,
    ROUND(SUM(transaction_amount) / SUM(transaction_count), 2) AS Avg_transaction_value
FROM agg_trans
GROUP BY Transaction_Type, state
ORDER BY Avg_transaction_value asc;
"""
dft4 = pd.read_sql(query4, mydb)
```

```python
dft4
```

# 2. User Engagement and Growth Strategy

```python
# 1 Total Registered Users and App Opens by State

query5 = """
SELECT 
    state,
    SUM(Registered_Users) AS Total_registered,
    SUM(App_Opens) AS Total_app_opens
FROM map_user
GROUP BY state
ORDER BY Total_registered desc;
"""
dft5 = pd.read_sql(query5, mydb)
```

```python
dft5
```

```python
# 2 State with High Registration but Low Engagement

query6 = """
SELECT 
    state,
    SUM(registered_users) AS Total_registered_users,
    SUM(app_opens) AS Total_app_opens,
    ROUND(SUM(app_opens) / NULLIF(SUM(registered_users), 0), 2) AS engagement_ratio,
    (SUM(app_opens) - SUM(registered_users)) AS gap
FROM map_user
GROUP BY state
ORDER BY Total_registered_users DESC;
"""
dft6 = pd.read_sql(query6, mydb)

# User engagement per person
# Finding underperformance relative to size
# Comparing big vs small states fairly
# Raw difference (e.g., how many users didn't open the app) gap
```

```python
dft6
```

```python
# 3 Total Registered Users and App Opens by State, distict

query7 = """
SELECT 
    state,
    district,
    SUM(Registered_Users) AS Total_registered_users,
    SUM(App_Opens) AS Total_app_opens
FROM map_user
GROUP BY state, district
ORDER BY Total_registered_users DESC;
"""
dft7 = pd.read_sql(query7, mydb)

# avg_percentage - share of current device type compared to all devices.
#  Average market share of that brand within the state (AVG(percentage))
```

```python
dft7
```

```python
# 4 bottom 10 registered user by state

query8 = """
SELECT 
    state,
    SUM(Registered_Users) AS Total_registered,
    SUM(App_Opens) AS Total_app_opens
FROM map_user
GROUP BY state
ORDER BY Total_registered asc
limit 10;
"""
dft8 = pd.read_sql(query8, mydb)

# focuse on low brand registerd users
```

# 3. Transaction Analysis Across States and Districts

```python
# 1. Top 10 States wise Transaction Value & Count

query9 = """
SELECT 
    state,
    SUM(transaction_count) AS Total_transactions,
    sum(transaction_amount) as Total_transaction_value
FROM top_trans
GROUP BY state
ORDER BY Total_transaction_value desc
limit 10;
"""
dft9 = pd.read_sql(query9, mydb)
```

```python
dft9
```

```python
# 2. top 10 District wise Transaction Value & Count

query10 = """
SELECT 
    Entity_Name as District,
    SUM(transaction_count) AS Total_transactions,
    sum(transaction_amount) as Total_transaction_value
FROM top_trans
where level = 'District'
GROUP BY District
ORDER BY Total_transaction_value desc
limit 10;
"""
dft10 = pd.read_sql(query10, mydb)
```

```python
dft10
```

```python
# 3. pincode wise Transaction Value & Count

query11 = """
SELECT 
    state,
    Entity_Name as Pincode,
    SUM(transaction_count) AS total_transactions,
    SUM(transaction_amount) AS total_value
FROM top_trans
WHERE level = 'pincode'
GROUP BY state, Entity_Name
ORDER BY total_value DESC;
"""
dft11 = pd.read_sql(query11, mydb)
```
