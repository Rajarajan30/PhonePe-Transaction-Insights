```python
import pandas as pd
import json
import os
```

```python
# this is to direct the path to get state data
# aggregated transaction

path = r"C:\Users\rajar\project phone pay\pulse\data\aggregated\transaction\country\india\state"
Agg_state_list_transaction = os.listdir(path)
Agg_state_list_transaction

# -----------------------------------------------------------------------

col_nam = {'State': [], 'Year' : [], 'Quarter' : [], 'Transaction_Type' : [], 'Transaction_Count' : [], 'Transaction_Amount' : []}

for i in Agg_state_list_transaction:
    p_i = os.path.join(path, i)          # path + i + '\\'
    Agg_yr = os.listdir(p_i)
    
    for j in Agg_yr:
        p_j = os.path.join(p_i, j)         # p_i + j + '\\'
        Agg_qu = os.listdir(p_j)
        
        for k in Agg_qu:
            p_k = os.path.join(p_j, k)     # p_j + k
            Data = open(p_k, 'r')  # read the json file
            D = json.load(Data) 
            Data.close()
            for z in D['data']['transactionData']:
                Name = z['name']
                Count = z['paymentInstruments'][0]['count']
                Amount = z['paymentInstruments'][0]['amount']
                
                col_nam['Transaction_Type'].append(Name)
                col_nam['Transaction_Count'].append(Count)
                col_nam['Transaction_Amount'].append(Amount)
                col_nam['State'].append(i)
                col_nam['Year'].append(j)
                col_nam['Quarter'].append(int(k.strip(' .json')))
# succesfully created data frame

Agg_data_transaction = pd.DataFrame(col_nam)                
```

```python
# aggregated user
# this is to direct the path to get state data

path = r"C:\Users\rajar\project phone pay\pulse\data\aggregated\user\country\india\state"
Agg_state_list = os.listdir(path)
Agg_state_list

# -----------------------------------------------------------------------

col_nam = {'State': [], 'Year' : [], 'Quarter' : [], 'Brand' : [], 'Transaction_Count' : [], 'Percentage' : []}

for i in Agg_state_list:
    p_i = os.path.join(path, i)          # path + i + '\\'
    Agg_yr = os.listdir(p_i)
    
    for j in Agg_yr:
        p_j = os.path.join(p_i, j)         # p_i + j + '\\'
        Agg_qu = os.listdir(p_j)
        
        for k in Agg_qu:
            p_k = os.path.join(p_j, k)     # p_j + k
            Data = open(p_k, 'r')  # read the json file
            D = json.load(Data) 
            Data.close()
            
            if D['data'] and D['data']['usersByDevice']:
                for z in D['data']['usersByDevice']:
                    Name = z['brand']
                    Count = z['count']
                    Amount = z['percentage']
                
                    col_nam['Brand'].append(Name)
                    col_nam['Transaction_Count'].append(Count)
                    col_nam['Percentage'].append(Amount)
                    col_nam['State'].append(i)
                    col_nam['Year'].append(j)
                    col_nam['Quarter'].append(int(k.strip(' .json')))
# succesfully created data frame

Agg_data_user = pd.DataFrame(col_nam)                
```

```python
# top

# top trancation
# this is to direct the path to get state data

path = r"C:\Users\rajar\project phone pay\pulse\data\top\transaction\country\india\state"
Agg_state_list_top_trancation = os.listdir(path)


# -----------------------------------------------------------------------

# col_nam = {'Year' : [], 'Quater' : [],'State': [], 'District' : [], 'Transaction_Count' : [], 'Transaction_Amount' : []}

rows = []

for i in Agg_state_list_top_trancation:
    p_i = os.path.join(path, i)          # path + i + '\\'
    Agg_yr = os.listdir(p_i)
    
    for j in Agg_yr:
        p_j = os.path.join(p_i, j)         # p_i + j + '\\'
        Agg_qu = os.listdir(p_j)
        
        for k in Agg_qu:
            p_k = os.path.join(p_j, k)     # p_j + k
            Data = open(p_k, 'r')  # read the json file
            D = json.load(Data) 
            Data.close()

            for level in ['states', 'districts', 'pincodes']:
                level_data = D['data'].get(level)    # geting level from loaded data
                if isinstance(level_data, list):    # avoid null error
                    for z in level_data:
                        rows.append({
                            'Year': j,
                            'Quarter': int(k.strip('.json')),
                            'State': i,
                            'Level': level[:-1].capitalize(),  # 'states' -> 'State'
                            'Entity_Name': z['entityName'],
                            'Transaction_Count': z['metric']['count'],
                            'Transaction_Amount': z['metric']['amount']
            })
# succesfully created data frame

Agg_data_top_trancation = pd.DataFrame(rows) 
```

```python
# creating table in db

table1 = """
CREATE TABLE IF NOT EXISTS agg_trans (
    State varchar(50),
    Year int,
    Quarter int,
    Transaction_Type varchar(40),
    Transaction_Count bigint,
    Transaction_Amount float

);
"""

curr.execute(table1)
db_connection.commit()
```
